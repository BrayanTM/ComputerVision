```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tablero = cv2.imread('../resources/img/tablero.jpg', 0)
plt.imshow(tablero, cmap = 'grey')
```

```python
found, esquinas = cv2.findChessboardCorners(tablero, (7, 7))
found
```

```python
ret, th1 = cv2.threshold(tablero, 160, 255, cv2.THRESH_BINARY)
plt.imshow(th1, cmap = 'grey')
```

```python
found, esquinas = cv2.findChessboardCorners(th1, (7, 7))
found
```

```python
cv2.drawChessboardCorners(th1, (7, 7), esquinas, found)
plt.imshow(th1)
```

# Puntos

```python
puntos = cv2.imread('../resources/img/puntos.jpg')
plt.imshow(puntos)
```

```python
found, esquinas = cv2.findCirclesGrid(puntos, (10, 10), cv2.CALIB_CB_SYMMETRIC_GRID)
found
```

```python
puntos_copia = puntos.copy()
cv2.drawChessboardCorners(puntos_copia, (10, 10), esquinas, found)
plt.imshow(puntos_copia)
```
