```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
halo = cv2.imread('../resources/img/HaloReach.jpg')
plt.imshow(halo)
```

```python
bordes = cv2.Canny(image = halo, threshold1 = 127, threshold2 = 127)
plt.imshow(bordes)
```

```python
bordes = cv2.Canny(image = halo, threshold1 = 0, threshold2 = 255)
plt.imshow(bordes)
```

## Formula para mejorar los resultados.

```python
mediana = np.median(halo)
lower = int(max(0, 0.7 * mediana))
upper = int(min(255, 1.3 * mediana))

bordes = cv2.Canny(image = halo, threshold1 = lower, threshold2 = upper)
plt.imshow(bordes)
```

## Aplicar un blur a la imagen puede mejorar aun mas el resultado.

```python
blur = cv2.blur(halo, ksize = (3, 3))

bordes = cv2.Canny(image = blur, threshold1 = lower, threshold2 = upper)
plt.imshow(bordes)
```
