```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('../resources/img/formas.png', 0)
plt.imshow(img, cmap = 'grey')
```

## Obtenemos los contornos y jerarquias de nuestra imagen.

```python
contornos, jerarquia = cv2.findContours(img, mode = cv2.RETR_CCOMP, method = cv2.CHAIN_APPROX_SIMPLE)
```

```python
type(contornos)
```

```python
len(contornos)
```

```python
jerarquia
```

## Obtenemos y Dibujamos los contornos externos.

```python
contornos_externos = np.zeros(img.shape)
```

```python
contornos_externos.shape
```

```python
for i in range(len(contornos)):
    if jerarquia[0][i][3] == -1: # Si jerarquia es igual a -1 son contornos externos.
        cv2.drawContours(contornos_externos, contornos, i, 255, -1)
```

```python
plt.imshow(contornos_externos, cmap = 'grey')
```

## Obtenemos y Dibujamos los contornos internos.

```python
contornos_internos = np.zeros(img.shape)
for i in range(len(contornos)):
    if jerarquia[0][i][3] != -1: # Si jerarquia no es igual a -1 son contornos internos.
        cv2.drawContours(contornos_internos, contornos, i, 255, -1)
plt.imshow(contornos_internos, cmap = 'grey')
```
