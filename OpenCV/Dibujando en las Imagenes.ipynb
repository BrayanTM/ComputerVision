```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

## Dibujamos una imagen con NumPy

```python
img = np.zeros(shape = (512, 512, 3), dtype = np.int16)
```

```python
plt.imshow(img)
```

## Dibujamos un cuadrado dentro de nuestra imagen con cv2

```python
cv2.rectangle(img, pt1 = (100, 100), pt2 = (200, 200), color = (236, 105, 45), thickness = 5)
plt.imshow(img)
```

```python
cv2.rectangle(img, pt1 = (300, 300), pt2 = (400, 400), color = (0, 2055, 0), thickness = 5)
plt.imshow(img)
```

## Dibujamos un circulo dentro de nuestra imagen.

```python
cv2.circle(img, center = (150, 350), radius = 50, color = (0, 0, 255), thickness = 5)
plt.imshow(img)
```

```python
cv2.circle(img, center = (150, 350), radius = 25, color = (0, 0, 255), thickness = 5)
plt.imshow(img)
```

```python
cv2.circle(img, center = (150, 350), radius = 75, color = (0, 0, 255), thickness = 5)
plt.imshow(img)
```

```python
cv2.circle(img, center = (350, 150), radius = 50, color = (0, 0, 255), thickness = -1)
plt.imshow(img)
```

## Dibujar una linea

```python
cv2.line(img, pt1 = (0, 0), pt2 = (512, 512), color = (2055, 0, 0), thickness = 5)
plt.imshow(img)
```

## Introducir Texto

```python
fuente = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img, text = 'Hola', org = (10, 500), fontFace = fuente, fontScale = 4, color = (255, 255, 255), thickness = 5, lineType = cv2.LINE_AA)
plt.imshow(img)
```

## Polgonos

```python
img = np.zeros(shape = (512, 512, 3), dtype = np.int16)
plt.imshow(img)
```

```python
vertices = np.array([[100, 300], [200, 200], [400, 300], [200, 400]], dtype = np.int32)
```

```python
vertices.shape
```

```python
puntos = vertices.reshape((-1, 1, 2))
```

```python
puntos.shape
```

```python
vertices
```

```python
puntos
```

```python
cv2.polylines(img, [puntos], isClosed = True, color = (236, 105, 45), thickness = 5)
plt.imshow(img)
```
