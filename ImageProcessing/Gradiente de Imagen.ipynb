```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap = 'grey')
```

```python
img = cv2.imread('../resources/img/crucigrama.jpg', 0)
```

```python
display_img(img)
```

### Vamos a hacer un sobel de x, esto nos permite ver las lineas verticales

```python
sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize = 5)
```

```python
display_img(sobelx)
```

### Vamos a hacer un sobel de y, esto nos permite ver las lineas horizontales

```python
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize = 5)
```

```python
display_img(sobely)
```

### Ahora vamos a crear un sobel de los dos x, y.

```python
sobelxy = cv2.addWeighted(src1 = sobelx, alpha = 0.5, src2 = sobely, beta = 0.5, gamma = 0)
```

```python
display_img(sobelxy)
```
