```python
img = cv2.imread('../resources/img/crucigrama.jpg', 0)
plt.imshow(img, cmap = 'grey')
```

```python
fig = plt.figure(figsize = (10, 10))
ax = fig.add_subplot(111)
ax.imshow(img, cmap = 'grey')
```

```python
corte2, thresh2 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
```

```python
fig = plt.figure(figsize = (10, 10))
ax = fig.add_subplot(111)
ax.imshow(thresh2, cmap = 'grey')
```

# Adaptive Threshold

```python
threshold3 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 8)
```

```python
fig = plt.figure(figsize = (10, 10))
ax = fig.add_subplot(111)
ax.imshow(threshold3, cmap = 'grey')
```

# Threshold y Threshold Adaptative Mezclados

```python
mezcla = cv2.addWeighted(src1 = thresh2, alpha = 0.5, src2 = threshold3, beta = 0.5, gamma = 0)
```

```python
fig = plt.figure(figsize = (10, 10))
ax = fig.add_subplot(111)
ax.imshow(mezcla, cmap = 'grey')
```
