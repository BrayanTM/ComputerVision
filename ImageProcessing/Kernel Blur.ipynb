```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('../resources/img/Kasumi.jpg').astype(np.float32)/255
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

**Gamma** nos permite hacer una imagen mas brillante o menos brillante.

```python
gamma = 1/4
gamma_img = np.power(img, gamma)
plt.imshow(gamma_img)
```

# Desenfoque
Agregamos Texto a la imagen para ver mejor el desenfoque.

```python
img = cv2.imread('../resources/img/Kasumi.jpg').astype(np.float32)/255
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
fuente = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(img, text = 'Kasumi', org = (175, 250), fontFace = fuente, fontScale = 2, color = (255, 255, 255), thickness = 3, lineType = cv2.LINE_AA)
plt.imshow(img)
```

```python
kernel = np.ones(shape = (5, 5), dtype=np.float32)/25
```

```python
kernel
```

```python
img_k = cv2.filter2D(img, -1, kernel)
```

```python
plt.imshow(img_k)
```

### Tambien podemos aplicar el blur o desenfoque de manera automatica con una funcion que nos brinda OpenCV

```python
img = cv2.imread('../resources/img/Kasumi.jpg').astype(np.float32)/255
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
fuente = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(img, text = 'Kasumi', org = (175, 250), fontFace = fuente, fontScale = 2, color = (255, 255, 255), thickness = 3, lineType = cv2.LINE_AA)
plt.imshow(img)
```

```python
img_b = cv2.blur(img, ksize = (5, 5))
plt.imshow(img_b)
```

### Otros tipos de blur

```python
img = cv2.imread('../resources/img/Kasumi.jpg').astype(np.float32)/255
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
fuente = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(img, text = 'Kasumi', org = (175, 250), fontFace = fuente, fontScale = 2, color = (255, 255, 255), thickness = 3, lineType = cv2.LINE_AA)
plt.imshow(img)
```

#### GaussianBlur

```python
img_b2 = cv2.GaussianBlur(img, (5, 5), 10)
plt.imshow(img_b2)
```

#### medianBlur

```python
img_b3 = cv2.medianBlur(img, 5)
plt.imshow(img_b3)
```
