```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
halo = cv2.imread('../resources/img/HaloReach.jpg')
halo_rgb = cv2.cvtColor(halo, cv2.COLOR_BGR2RGB)

kasumi = cv2.imread('../resources/img/Kasumi.jpg')
kasumi_rgb = cv2.cvtColor(kasumi, cv2.COLOR_BGR2RGB)

zero = cv2.imread('../resources/img/Zero.jpg')
zero_rgb = cv2.cvtColor(zero, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(zero_rgb)
```

# Histograma

### Rojo

```python
# OpenCV es BGR
hist = cv2.calcHist([zero], channels = [2], mask = None, histSize = [256], ranges = [0, 256])
```

```python
plt.plot(hist)
```

### Verde

```python
# OpenCV es BGR
hist2 = cv2.calcHist([zero], channels = [1], mask = None, histSize = [256], ranges = [0, 256])
```

```python
plt.plot(hist2)
```

Plotear los canales en un mismo grafico.

```python
img = kasumi
color = ('b', 'g', 'r')
for i, col in enumerate(color):
    histr = cv2.calcHist([img], [i], None, [256], [0, 256])
    plt.plot(histr, color = col)
    plt.xlim([0, 256])
plt.title('Kasumi')
plt.show()
```
