```python
import numpy as np
from numpy import genfromtxt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense
from sklearn.metrics import confusion_matrix, classification_report
```

https://archive.ics.uci.edu/dataset/267/banknote+authentication

```python
data = genfromtxt('../resources/data/bank_note_data.txt', delimiter=',')
```

```python
data
```

```python
etiquetas = data[:,4]
etiquetas
```

```python
caracteristicas = data[:,0:4]
caracteristicas
```

```python
X = caracteristicas
y = etiquetas
```

## Sepramos los datos en train y test

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
len(X_train)
```

```python
len(X)
```

```python
len(X_test)
```

```python
len(y_train)
```

```python
len(y_test)
```

### Normalizar los datos

```python
scaler_object = MinMaxScaler()
```

```python
scaler_object.fit(X_train)
```

```python
scaled_X_train = scaler_object.transform(X_train)
```

```python
scaled_X_train.max()
```

```python
scaled_X_test = scaler_object.transform(X_test)
```

```python
scaled_X_test.max()
```

## Construimos el modelo

```python
model = Sequential()
model.add(Dense(4, input_dim = 4, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

## Entrenamos el Modelo

```python
model.fit(scaled_X_train, y_train, epochs=100, verbose=2)
```

```python
predictions = model.predict(scaled_X_test)
```

```python
classes = (predictions > 0.5).astype(int)
```

```python
classes
```

## Evaluamos los resultados

```python
model.metrics_names
```

```python
model.evaluate(x = scaled_X_test, y = y_test)
```

```python
confusion_matrix(y_test, classes)
```

```python
print(classification_report(y_test, classes))
```

## Guardar el modelo

```python
model.save('mi_modelo.h5')
```
