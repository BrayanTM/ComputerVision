## RGB - Cambiando los Canales de una imagen
### 0 a 255

```python
img_rojo.shape
```

```python
# Seleccionamos el canal Rojo
img_rojo[:, :, 0]
```

```python
# Mostramos el canal seleccionado con Matplotlib.
# Por defecto matplotlib utiliza la coloracion viridis por lo que la imagen no se mostrara como esperamos.
plt.imshow(img_rojo[:, :, 0])
```

```python
plt.imshow(img_rojo[:, :, 0], cmap = 'grey')
```

```python
plt.imshow(img_rojo[:, :, 1], cmap = 'grey')
```

```python
plt.imshow(img_rojo[:, :, 2], cmap = 'grey')
```

```python
# Establecemos en cero el canal Verde
img_rojo[:, :, 1] = 0
```

```python
# Mostramos el resultado.
plt.imshow(img_rojo)
```

```python
# Establecemos el color azul en cero y mostramos la imagen.
img_rojo[:, :, 2] = 0
plt.imshow(img_rojo)
```
