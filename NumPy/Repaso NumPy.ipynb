## Operaciones

**random.seed()**: Indica la semilla por defecto al generar un valor o conjunto de valores random.

```python
np.random.seed(33)
```

```python
array = np.random.randint(0, 100, 10)
print(array)
```

Max devuelve el valor maximo dentro de un array, mientras que argmax devuelve la posicion donde se encuentra ese valor.

Aplica igual mar min y argmin.

```python
array.max()
```

```python
array.argmax()
```

```python
array.min()
```

```python
array.argmin()
```

`Mean` devuelve la media en un arreglo, `reshape` cambia la forma del arreglo.

```python
array.mean()
```

```python
array.reshape(2, 5)
```
